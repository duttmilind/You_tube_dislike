. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'
and so on.....

```python
def chec(month):
    
        
        if(month==1):
        
            return 'jan'
        elif month == 2:
            return 'feb'
        elif month == 3:
            return 'mar'
        elif month == 4:
            return 'apr'
        elif month == 5:
            return 'may'
        elif month == 6:
            return 'June'
        elif month == 7:
            return 'July'
        elif month == 8:
            return 'Agu'
        elif month == 9:
            
            return 'Sep'
        elif month == 10:
            return 'Oct'
        elif month == 11:
            return 'Nov'
        elif month == 12:
            return 'Dec'
        
        else:
            return 'unknown'
```

```python
df2['published_month']=df2['published_month'].apply(chec)
```

```python
df2['published_month']
```

Inference

 the numbers in the column published_month replaced as names of the months
